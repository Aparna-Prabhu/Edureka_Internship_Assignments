```python
#Problem Statement 2: Write a Python program to perform the following tasks-
#1.Create a pandas series from the below dictionary where indices are subjects:{'English':{'Sam':60,'Jackson':74,'Ahree':85}, 'History':{'Gloria':83,'Sam':65,'Isla':78,'Aron':72,'Gray':61},'Geography':{'Jackson':92,'Gloria':95,'Isla':82,'Aron':75,'Ahree':76},'Mathematics':{'Sam':99,'Gloria':74,'Jackson':89,'Ahree':85,'Gray':95},'Science':{'Sam':89,'Aron':82,'Gray':78,'Isla':93,'Ahree':87}}
#2.Convert the created series into DataFrame and replace the null values with zeroes.
#3.Transpose the DataFrame and create a new column 'Average' and fill the values in it by calculating the average of all subjects.
```

```python
#TASK 1
value= {'English':{'Sam':60,'Jackson':74,'Ahree':85}, 
        'History':{'Gloria':83,'Sam':65,'Isla':78,'Aron':72,'Gray':61},
        'Geography':{'Jackson':92,'Gloria':95,'Isla':82,'Aron':75,'Ahree':76},
        'Mathematics':{'Sam':99,'Gloria':74,'Jackson':89,'Ahree':85,'Gray':95},
        'Science':{'Sam':89,'Aron':82,'Gray':78,'Isla':93,'Ahree':87}}
s = pd.Series(value)
s= s.apply(pd.Series)
s
```

```python
#Task 2
df=pd.DataFrame(s)
df=df.fillna(0)
df
```

```python
#3.Transpose the DataFrame and create a new column 'Average' and 
#fill the values in it by calculating the average of all subjects.
df=df.transpose()
df
```

```python
#fill the values in it by calculating the average of all subjects.
new_col = df["Average"] = df.mean(numeric_only = True, axis = 1)
new_col
```

```python
#Problem Statement 3: Write a Python program to create a series from 1 to 1000 and select only numbers divisible by 7 and 17.
#num = pd.Series(range(1,1000)
num_list = []
for i in range(1,1000): 
    if i % 7 == 0 or i % 17 == 0: 
      num_list.append(i)

num_series = pd.Series(num_list)
num_series
```

```python
Problem Statement 4: Sylphia has a dataset of various cereals sold in the supermarket.
Dataset Description:cereal.csv-The dataset contains 16 features. 
Here's a brief description of 3 columns in the dataset:
•name-Brand name of the cereals
•MFR-Manufacturer of the brands
•rating-Rating of the cerealsSyliphia wants to visualize the quality of cereals and determine which manufacturer delivers the best quality.
Tasks to be performed:  
Fetch and download the dataset using the following link https://www.dropbox.com/s/idnul34dfo5cnke/cereal.csv
```

```python
#1.Import the dataset.
df = pd.read_csv(r"C:\Users\Admin\Desktop\edureka_demos\cereal.csv")
value = df
value
```

```python
#2.Plot ratings of different types of manufacturers.
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.figure(figsize=(10,8))
for group,data in value.groupby("mfr"):
    plt.barh(group,data["rating"])
```

```python
#3.Use xticks range form 0-100
for group,data in value.groupby("mfr"):
    plt.barh(group,data["rating"])
plt.xticks(np.arange(0,100,10))
plt.show()
```

```python
#4.Change the style of the graph to seaborn:
import seaborn as sns
import warnings
f,ax = plt.subplots()

sns.barplot(x ='rating', y = 'mfr', data = value,ax = None).set(xlabel = 'ratings', ylabel = 'Manufacturers')
plt.show()
```
